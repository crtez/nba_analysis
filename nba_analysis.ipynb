```python
import pandas as pd
```

```python
df = pd.read_csv("games.csv")
df['GAME_DATE_EST'] = pd.to_datetime(df['GAME_DATE_EST'])
df['total_points'] = df['PTS_home'] + df['PTS_away']
```

```python
df['SEASON'].unique()
```

```python
grouped_df = df.groupby(by = "SEASON").mean()
grouped_df
```

```python
grouped_df.plot(y = "total_points", xlim = [2003, 2021], 
                grid = True, ylabel = "Average points per game (home + away)", 
                title = "NBA Score Inflation (2003 - 2021)")
```
